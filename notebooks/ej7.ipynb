```python
# Solicite al usuario una lista de nombres de participantes (separados por
# coma). El programa debe asignar aleatoriamente a cada persona un amigo
# invisible, cumpliendo que nadie se tenga a sí mismo.
# Validaciones:
# Debe haber al menos 3 participantes.
# No debe haber nombres duplicados (comparar ignorando mayúsculas y
# espacios extra).

import random

participants = input("Ingrese una serie de nombres de participantes (separados por coma): ")
names = [name.strip() for name in participants.split(",")]

if len(names) < 3:
    print("Debe haber al menos 3 participantes")
    correct = False
else:
    correct = True

if correct:
    names_lower = [name.lower() for name in names]

    if len(names_lower) != len(set(names_lower)):
        print("No se permiten nombres duplicados!.")
        correct = False
    else:
        correct = True

    if correct:
        mix = names[:]
        random.shuffle(mix)
        invisible_friend = {}

        for i in range(len(mix)):
            person = mix[i]
            friend = mix[(i + 1) % len(mix)]
            invisible_friend[person] = friend
    
        print("Sorteo de amigo invisible: ")

        for person, friend in invisible_friend.items():
            print(f"{person} → {friend}")

    

```
